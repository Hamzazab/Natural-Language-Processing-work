```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
#apply stopword remove sapces , dot,- , regular expresion etc
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

ps = PorterStemmer()

corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z\s.-]', ' ' , messages['text'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
print(corpus[3])
```

```python
#Apply bag of words
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000 , ngram_range=(1,3))
x = tfidf.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
y = messages['label']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
# Access the feature names from the CountVectorizer's vocabulary_
feature_names = tfidf.get_feature_names_out()

# Print the first 20 feature names
print(feature_names[:20])
```

```python
classifier.fit(x_train,y_train)
```

```python
import numpy as np
previous_score=0
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(x_train,y_train)
    y_pred=sub_classifier.predict(x_test)
    acc_score = accuracy_score(y_test,pred)
    if acc_score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,acc_score))
```

```python

# Assuming you are interested in the most relevant words for class 0 (negative class)
class_index = 1

# Access the feature log probabilities for the class of interest
feature_log_probabilities = classifier.feature_log_prob_

# Get the feature names from the CountVectorizer
feature_names = tfidf.get_feature_names_out()

# Get the indices of the words with the highest feature log probabilities for the class
top_word_indices = np.argsort(feature_log_probabilities[class_index])[::-1]

# Show the top N words (e.g., top 10) with the highest feature log probabilities
top_N = 20
top_words = [feature_names[i] for i in top_word_indices[:top_N]]

print(f"Most relevant words for class {class_index}:")
print(top_words)
```

```python
# Assuming you are interested in the most irrelevant words for class 0 (negative class)
class_index = 1

# Access the feature log probabilities for the class of interest
feature_log_probabilities = classifier.feature_log_prob_

# Get the feature names from the CountVectorizer
feature_names = tfidf.get_feature_names_out()

# Get the indices of the words with the most negative feature log probabilities for the class
top_word_indices = np.argsort(feature_log_probabilities[class_index])

# Show the top N words (e.g., top 10) with the most negative feature log probabilities
top_N = 5000
top_words = [(feature_names[i], feature_log_probabilities[class_index][i]) for i in top_word_indices[:top_N]]

print(f"Most irrelevant words for class {class_index}:")
for word, coefficient in top_words:
    print(f"Coefficient: {coefficient:.4f}, Word: {word}")
```
